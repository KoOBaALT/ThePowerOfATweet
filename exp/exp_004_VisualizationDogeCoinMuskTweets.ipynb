## Code to Plot DogeCoin Value and Elon Musks Tweets

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.dates as mdates
from tueplots import bundles


df_dog = pd.read_csv("../dat/crypto-stocks/doge-coin-daily.csv")
df_merged = pd.read_csv("../dat/cleaned_data/hot_tweets_merged_dataset.csv")
df_tweets = df_merged [~df_merged.tweet.isnull()]
df_only_tweets = df_tweets [df_tweets.tweet.str[0] != "@"]
df_only_tweets = df_only_tweets [~df_only_tweets.tweet.duplicated()]
tweet_dates = df_only_tweets.Date.unique()


plt.rcParams.update(bundles.neurips2021(usetex=False))

day_delay = 7

fig, ax = plt.subplots(figsize=[20,10])
ax.plot(pd.to_datetime(df_dog.Date),df_dog["Close"], color ="gray")
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.set_xlabel("Date")
ax.set_ylabel("DogeCoin Price")



# First Tweet
ax.vlines(x = pd.to_datetime("2019-04-02"), ymin=-0.01,ymax=0.7, color="black", alpha=0.8, linestyles="dashed")

ix = df_dog [df_dog.Date == tweet_dates[0]].index[0]
df_plot = df_dog.iloc[ix-day_delay:ix]
df_plot = df_plot.append(df_dog.iloc[ix:ix+day_delay])
tweets_used = df_only_tweets [df_only_tweets.Date == tweet_dates[0]].tweet.values

ax2 = plt.axes([.08, .2, .2, .2])
ax2.plot(pd.to_datetime(df_plot.Date),df_plot["Close"], color="gray")
ax2.set_xlim(pd.to_datetime([pd.to_datetime(df_plot.Date).iloc[0], pd.to_datetime(df_plot.Date).iloc[-1]]))
ax2.axvline(x=pd.to_datetime(tweet_dates[0]),color="black", alpha=0.8)
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax2.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax2.set_title("1# Tweet Day\nTweet: "+"\nTweet: ".join(tweets_used))

x1 = [pd.to_datetime(tweet_dates[0]), pd.to_datetime(tweet_dates[0])+pd.Timedelta(days=29)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.085]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

x1 = [pd.to_datetime(tweet_dates[0]), pd.to_datetime(tweet_dates[0])+pd.Timedelta(days=243)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.085]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)


# Second Tweet
ax.vlines(x = pd.to_datetime("2021-02-10"), ymin=-0.01,ymax=0.7, color="black", alpha=0.8, linestyles="dashed")

ix = df_dog [df_dog.Date == tweet_dates[1]].index[0]
df_plot = df_dog.iloc[ix-day_delay:ix]
df_plot = df_plot.append(df_dog.iloc[ix:ix+day_delay])
tweets_used = df_only_tweets [df_only_tweets.Date == tweet_dates[1]].tweet.values

ax3 = plt.axes([.35, .12, .2, .2])
ax3.plot(pd.to_datetime(df_plot.Date),df_plot["Close"], color="gray")
ax3.set_xlim(pd.to_datetime([pd.to_datetime(df_plot.Date).iloc[0], pd.to_datetime(df_plot.Date).iloc[-1]]))
ax3.axvline(x=pd.to_datetime(tweet_dates[1]), color="black", alpha=0.8)
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax3.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax3.set_title("2# Tweet Day\nTweet: "+"\nTweet: ".join(tweets_used))

x1 = [pd.to_datetime(tweet_dates[1]), pd.to_datetime(tweet_dates[1])-pd.Timedelta(days=140)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.183]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

x1 = [pd.to_datetime(tweet_dates[1]), pd.to_datetime(tweet_dates[1])-pd.Timedelta(days=140)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.0195]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)


# Third Tweet
ax.vlines(x = pd.to_datetime("2021-02-04"), ymin=-0.01,ymax=0.7, color="black", alpha=0.8, linestyles="dashed")

ix = df_dog [df_dog.Date == tweet_dates[2]].index[0]
df_plot = df_dog.iloc[ix-day_delay:ix]
df_plot = df_plot.append(df_dog.iloc[ix:ix+day_delay])
tweets_used = df_only_tweets [df_only_tweets.Date == tweet_dates[2]].tweet.values

ax3 = plt.axes([.4, .5, .2, .2])
ax3.plot(pd.to_datetime(df_plot.Date),df_plot["Close"], color="gray")
ax3.set_xlim(pd.to_datetime([pd.to_datetime(df_plot.Date).iloc[0], pd.to_datetime(df_plot.Date).iloc[-1]]))
ax3.axvline(x=pd.to_datetime(tweet_dates[2]), color="black", alpha=0.8)
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax3.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax3.set_title("3# Tweet Day\nTweet: "+"\nTweet: ".join(tweets_used))

x1 = [pd.to_datetime(tweet_dates[2]), pd.to_datetime(tweet_dates[2])-pd.Timedelta(days=150)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.492]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

x1 = [pd.to_datetime(tweet_dates[2]), pd.to_datetime(tweet_dates[2])-pd.Timedelta(days=307)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.328]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)


# Fouth Tweet
ax.vlines(x = pd.to_datetime("2021-02-14"), ymin=-0.01,ymax=0.7, color="black", alpha=0.8, linestyles="dashed")

ix = df_dog [df_dog.Date == tweet_dates[3]].index[0]
df_plot = df_dog.iloc[ix-day_delay:ix]
df_plot = df_plot.append(df_dog.iloc[ix:ix+day_delay])
tweets_used = df_only_tweets [df_only_tweets.Date == tweet_dates[3]].tweet.values
tweets_used[0] = tweets_used[0][0:71] + "\n"+tweets_used[0][71:]

ax3 = plt.axes([.74, .7, .2, .2])
ax3.plot(pd.to_datetime(df_plot.Date),df_plot["Close"], color="gray")
ax3.set_xlim(pd.to_datetime([pd.to_datetime(df_plot.Date).iloc[0], pd.to_datetime(df_plot.Date).iloc[-1]]))
ax3.axvline(x=pd.to_datetime(tweet_dates[3]), color="black", alpha=0.8)
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax3.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax3.set_title("4# Tweet Day\nTweet: "+"\nTweet: ".join(tweets_used),bbox=dict(facecolor='white',edgecolor='none', alpha=1))

x1 = [pd.to_datetime(tweet_dates[3]), pd.to_datetime(tweet_dates[3])+pd.Timedelta(days=55)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.49]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

x1 = [pd.to_datetime(tweet_dates[3]), pd.to_datetime(tweet_dates[3])+pd.Timedelta(days=270)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.49]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)


# Fith Tweet
ax.vlines(x = pd.to_datetime("2021-03-13"), ymin=-0.01,ymax=0.7, color="black", alpha=0.8, linestyles="dashed")

ix = df_dog [df_dog.Date == tweet_dates[4]].index[0]
df_plot = df_dog.iloc[ix-day_delay:ix]
df_plot = df_plot.append(df_dog.iloc[ix:ix+day_delay])
tweets_used = df_only_tweets [df_only_tweets.Date == tweet_dates[4]].tweet.values
tweets_used[0] = tweets_used[0][0:61] + "\n"+tweets_used[0][61:128]+"\n"+tweets_used[0][128:]
ax3 = plt.axes([.78, .1, .2, .2])
ax3.plot(pd.to_datetime(df_plot.Date),df_plot["Close"], color="gray")
ax3.set_xlim(pd.to_datetime([pd.to_datetime(df_plot.Date).iloc[0], pd.to_datetime(df_plot.Date).iloc[-1]]))
ax3.axvline(x=pd.to_datetime(tweet_dates[4]), color="black", alpha=0.8)
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax3.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax3.set_title("5# Tweet Day\nTweet: "+"\nTweet: ".join(tweets_used),bbox=dict(facecolor='white',edgecolor='none', alpha=1))


x1 = [pd.to_datetime(tweet_dates[4]), pd.to_datetime(tweet_dates[4])+pd.Timedelta(days=69)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.165]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

x1 = [pd.to_datetime(tweet_dates[4]), pd.to_datetime(tweet_dates[4])+pd.Timedelta(days=69)]
y1 = [df_plot.iloc [int(len(df_plot)/2),:].Close, 0.004]
ax.plot(x1, y1, marker = '', color ="gray", alpha = 0.25)

#ax.set_title("DogeCoin Timeseries with Elon Musks Tweets about DogeCoin", fontsize=28)
ax.set_xlim(pd.to_datetime(["2019-03-01", "2022-01-12"]))
plt.show()
```
