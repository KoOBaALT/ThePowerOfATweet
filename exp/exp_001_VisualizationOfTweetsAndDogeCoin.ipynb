### Visualization Of Tweets And DogeCoin

Aim: Create a nice plotly visualization of the DogeCoin timeseries and Elon Musks Tweets about DogeCoins

```python
import pandas as pd
import plotly.express as px

df_doge = pd.read_csv("../dat/crypto-stocks/doge-coin-daily.csv")
df_doge["OHLC Average"] = df_doge.iloc[:,1:5].mean(axis = 1)
df_doge.Date = pd.to_datetime(df_doge.Date)
```

```python
!pip install plotly
import plotly.express as px
fig = px.line(df_doge, x="Date", y="high", title='DogeCoin Price')
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.line(df_doge, x="Date", y="OHLC Average", title='DogeCoin Price')

fig.add_trace(go.Scatter(
    x=["2020-09-24", "2021-09-24"],
    y=[1, 1],
    mode="markers+text",
    name="Tweets",
    text=["Text A", "Text B"],
    textposition="top center"
))

#fig.add_vline(x='2020-09-24')
fig.show()
```
